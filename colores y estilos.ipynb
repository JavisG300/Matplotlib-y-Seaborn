```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
fig, axes = plt.subplots(figsize = (8,8))
```

```python
plt.style.use('dark_background')
fig, axes = plt.subplots(figsize = (8,8))
```

```python
plt.style.use('Solarize_Light2')
fig, axes = plt.subplots(figsize = (8,8))
```

```python
plt.style.use('dark_background')
fig, axes = plt.subplots(figsize = (8,8))
axes.plot(x,x+1,'r')
axes.plot(x,x+2,'b')
axes.plot(x,x+3,'y')
axes.plot(x,x+4,'g')
```

Podemos hacer modificaciones de tipo **MATLAB**

```python
plt.style.use('dark_background')
fig, axes = plt.subplots(figsize = (8,8))
axes.plot(x,x+1,'r--')
axes.plot(x,x+2,'b:')
axes.plot(x,x+3,'yD')
axes.plot(x,x+4,'go-')
```

**Pyplot**

```python
plt.style.use('dark_background')
fig, axes = plt.subplots(figsize = (8,8))
axes.plot(x,x+1,color = 'green', alpha = 0.5,linewidth = 5,linestyle = '-',marker = 'o') #alpha me da un nivel de transparencia, linewidth para el grosor. linestyle
axes.plot(x,x+2,color = 'blue',linewidth = 3,linestyle = '--',marker = 'x',markersize = 20) #Marker y marker size son las figuras y su tamaño
axes.plot(x,x+3,color = '#66FF66',linewidth = 10,linestyle = ':',marker = 'v', markerfacecolor = '#CCFFCC') #Colores en modo rgb
axes.plot(x,x+4,color = '#CCFFCC',linewidth = 2,linestyle = 'dashed',marker = 'P')
```
