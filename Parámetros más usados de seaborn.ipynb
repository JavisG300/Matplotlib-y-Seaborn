```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Seaborn tiene datasets predeterminados

```python
tips = sns.load_dataset('tips') #Data set de propinas
```

```python
tips.head()
```

Seaborn determina de manera autom치tica el gr치fico a usar 

```python
sns.displot(data = tips, x = 'total_bill')
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill', y = 'tip')
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill', y = 'tip',hue='sex')
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill',hue='sex') #Puede hacer la distinci칩n en gr치fico de una sola variable
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill',hue='sex',kind='kde') #kind permite elegir el tipo de gr치fico
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill',hue='sex',kind='kde', legend=False)
plt.show()
```

```python
sns.displot(data = tips, x = 'total_bill',hue='sex',kind='kde', legend=False,palette = 'dark', alpha = 0.25)
plt.show()
```
