Un gráfico de barras nos permite graficar variables categóricas, es decir variables de texto

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN','MEXICO', 'GERMANY', 'COLOMBIA']
population = [1000,800,450,900,2500]
```

```python
plt.bar(country,population)
plt.show()
```

Width

```python
plt.bar(country,population,width = 0.5)
plt.show()
```

Color

```python
plt.bar(country,population,width = 0.5, color = ['red','green'])
plt.show()
```

Supongamos que quiero cambiar el nombre de una categoría

```python
plt.bar(country,population,width = 0.5, color = ['red','green'])
plt.xticks(np.arange(5),('India', 'Japón', 'México','Alemania','Colombia'))
plt.show()
```

¿Y si quiero que los nombres no estén amontonados?

```python
plt.bar(country,population,width = 0.5, color = ['red','green'])
plt.xticks(np.arange(5),('India', 'Japón', 'México','Alemania','Colombia'),rotation = 45)
plt.show()
```

Dandole estilo

```python
plt.style.use('ggplot')
plt.bar(country,population,width = 0.5, color = ['#355070','#6d597a','#b56576','#e56b6f','#eaac8b'])
plt.xticks(np.arange(5),('India', 'Japón', 'México','Alemania','Colombia'),rotation = 45)
plt.title('Pobalción por país')
plt.xlabel('País')
plt.ylabel('Población')
plt.show()
```
