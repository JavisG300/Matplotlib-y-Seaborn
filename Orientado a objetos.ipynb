```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

Primero hay que crear una figura, esta figura es el lienzo sobre el que estará nuestra gráfica

```python
fig = plt.figure()
```

¿Cuántas gráficas podemos "pintar"? las que queramos, nos lo permite. Dichas gráficas se reconocen como **axes**

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9]) #Los dos primeros número hacen referencia a la posición en la figura(lienzo)
#Los siguientes 2 hacen referencia al tamaño
axes.plot(x,y,'b')
fig.show()
```

Pero, ¿por qué complicarnos con esto? si con pyplot ya lo podemos hacer


La ventaja es si queremos hacer algo mucho más epecífico

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r--')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r--')
axes2.set_facecolor('gray') #Cada objeto es distinto y lo puedo editar de manera individual
fig.show()
```

Esta forma de hacerlo, orientado a objetos, permite hacer modifcaciones específicas a cada una de las figuras
