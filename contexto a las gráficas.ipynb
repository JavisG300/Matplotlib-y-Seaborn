```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

Así están muy simples

### Poniendo **título**

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('Relación x - y') #Fijando el título
axes[1].plot(y,x)
axes[1].set_title('Relación y - x')
```

### Poniendo etiquetas a los **ejes**

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[0].set_title('Relación x - y') #Fijando el título
axes[0].set_xlabel('x')
axes[0].set_ylabel('y = sin(x)')
axes[1].plot(y,x)
axes[1].set_title('Relación y - x')
axes[1].set_xlabel('y = sin(x)')
axes[1].set_ylabel('x')
```

### Cambiando el tamaño de la figura

```python
fig, axes = plt.subplots(1,2, figsize = (10,5))
axes[0].plot(x,y)
axes[0].set_title('Relación x - y') #Fijando el título
axes[0].set_xlabel('x')
axes[0].set_ylabel('y = sin(x)')
axes[1].plot(y,x)
axes[1].set_title('Relación y - x')
axes[1].set_xlabel('y = sin(x)')
axes[1].set_ylabel('x')
```

### Uso de labels

```python
fig, axes = plt.subplots(1,2, figsize = (10,5))
axes[0].plot(x,y,label = '$sin(x)$') #$$ como en latex
axes[0].set_title('Relación x - y') #Fijando el título
axes[0].set_xlabel('x')
axes[0].set_ylabel('y = sin(x)')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relación y - x')
axes[1].set_xlabel('y = sin(x)')
axes[1].set_ylabel('x')
```

## TRANQUILOS, se puede hacer sin el método orientado a objetos

```python
plt.plot(x,y,label = 'sin(x)')
plt.title('Título')
plt.xlabel('x')
plt.ylabel('y = sin(x)')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.5,0.2)) #Puedo decirle donde poner la leyenda
```
