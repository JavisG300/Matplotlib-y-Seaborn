```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

## Histograma

```python
plt.hist(data)
plt.show()
```

#### Bins

En un histograma **bins** es el número de barras en el cuál va a llegar la distribución de datos

```python
plt.hist(data,bins = 25)
plt.show()
```

#### Histtype

```python
plt.hist(data,bins = 25,histtype='step')
plt.show()
```

## Boxplot

Cajas y bigotes: muestra los rangos percentilicos 25,50,75 y 100

```python
plt.boxplot(data)
plt.show()
```

Algunas propiedades

```python
data = np.append(data,200) #Datos atípicos
plt.boxplot(data,vert=False,patch_artist=True,notch=True,showfliers=True) #showflires es para ver los datos anómalos
plt.show()
```

## Scatter

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (30 * x)**2
colors = y
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s=area)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='D')
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='D',alpha=0.5)
plt.show()
```
